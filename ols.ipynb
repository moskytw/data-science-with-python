```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import IPython as ip
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
mpl.style.use('ggplot')
mpl.rc('figure', figsize=(7.2, 5.76))
mpl.rc('font', family='Noto Sans CJK TC')
plt.rc('lines', markeredgecolor='white', markeredgewidth=0.75)
plt.rc('patch', edgecolor='white', force_edgecolor=True, linewidth=1)
ip.display.set_matplotlib_formats('svg')
```

```python
m = sm.datasets.anes96
df = m.load_pandas().data
```

```python
smf.ols('vote ~ PID', df).fit().summary()
```

```python
smf.ols('vote ~ selfLR', df).fit().summary()
```

```python
smf.ols('vote ~ ClinLR', df).fit().summary()
```

```python
smf.ols('vote ~ income', df).fit().summary()
```

```python
smf.ols('vote ~ logpopul', df).fit().summary()
```

```python
smf.ols('vote ~ educ', df).fit().summary()
```

```python
smf.ols('vote ~ age', df).fit().summary()
```

```python
# smf.ols('vote ~ PID + selfLR', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + income', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + logpopul', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + educ', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + age', df).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR', df).fit().summary()
smf.ols('vote ~ 0 + PID + selfLR + ClinLR', df).fit().summary()
```

We can say: “After control the variables, the _income_ and the other variables becomes insignificant.”

```python
res = smf.ols('vote ~ 0 + PID + selfLR + ClinLR', df).fit()
res.predict(dict(PID=2, selfLR=3, ClinLR=2))
```
