```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import IPython as ip
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
mpl.style.use('ggplot')
mpl.rc('figure', figsize=(7.2, 5.76))
mpl.rc('font', family='Noto Sans CJK TC')
plt.rc('lines', markeredgecolor='white', markeredgewidth=0.75)
plt.rc('patch', edgecolor='white', force_edgecolor=True, linewidth=1)
ip.display.set_matplotlib_formats('svg')
```

```python
m = sm.datasets.anes96
df = m.load_pandas().data
df_raw = df
```

```python
smf.ols('vote ~ PID', df_raw).fit().summary()
```

```python
smf.ols('vote ~ selfLR', df_raw).fit().summary()
```

```python
smf.ols('vote ~ ClinLR', df_raw).fit().summary()
```

```python
smf.ols('vote ~ income', df_raw).fit().summary()
```

```python
smf.ols('vote ~ logpopul', df_raw).fit().summary()
```

```python
smf.ols('vote ~ educ', df_raw).fit().summary()
```

```python
smf.ols('vote ~ age', df_raw).fit().summary()
```

```python
# smf.ols('vote ~ PID + selfLR', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + income', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + logpopul', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + educ', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR + age', df_raw).fit().summary()
# smf.ols('vote ~ PID + selfLR + ClinLR', df_raw).fit().summary()
# smf.ols('vote ~ 0 + PID + selfLR + ClinLR', df_raw).fit().summary()
res = smf.ols('vote ~ 0 + PID + selfLR + ClinLR', df_raw).fit()
display(res.summary())
res_ols_1 = res
```

We can say: “After control the variables, the _income_ and the other variables becomes insignificant.”

```python
s = res_ols_1.params
s = s.sort_values()
s.plot.bar()
```

```python
res_ols_1.predict(dict(PID=2, selfLR=3, ClinLR=2))
```
