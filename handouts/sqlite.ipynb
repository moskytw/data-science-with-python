```python
import sqlite3
import pandas as pd
import statsmodels.api as sm
```

```python
conn = sqlite3.connect('inputs/anes96.db')
```

```python
m = sm.datasets.anes96
df = m.load_pandas().data
display(df)
df_raw = df
```

```python
# `'anes96'` is the name of table
df_raw.to_sql('anes96', index=False, if_exists='replace', con=conn)
```

```python
pd.read_sql('select * from anes96', con=conn)
```
