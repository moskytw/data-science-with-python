# Standardization

How to compare the different distribution? Standardize them!

$ z = {\dfrac {x-\mu }{\sigma }} $

* $ z $: z-score = standard score.
* $ x $: raw score.
* $ \mu $: mean.
* $ \sigma $: standard deviation.

<img style="float: left; width: 50%;" alt="Normal_distribution_and_scales.gif" src="https://upload.wikimedia.org/wikipedia/commons/b/bb/Normal_distribution_and_scales.gif">
<div style="clear: both"></div>

* t-score:
    * In hypothesis testing, t-score is the test statistic follows a Student's t-distribution under the null hypothesis.
    * In educational assessment, t-score is the z-score shifted and scaled to have a mean of 50 and a standard deviation of 10.
    * In bone density measurements, t-score is the z-score of the measurement compared to the population of healthy 30-year-old adults.

```python
howell1_df = pd.read_csv('dataset_howell1.csv', sep=';')
```

## By Hand

```python
# it's weird to compare height and weight,
# but ... anyway it's just an example
sns.distplot(howell1_df.height)
sns.distplot(howell1_df.weight)
```

### Mean Removal

Make the mean to zero.

```python
sns.distplot(howell1_df.height - howell1_df.height.mean())
sns.distplot(howell1_df.weight - howell1_df.weight.mean())
```

### Variance Scaling

Make the variance to unit variance.

```python
# we can still find the height is more centralized (higher kurtosis)
sns.distplot((howell1_df.height - howell1_df.height.mean()) / howell1_df.height.std())
sns.distplot((howell1_df.weight - howell1_df.weight.mean()) / howell1_df.weight.std())
```
