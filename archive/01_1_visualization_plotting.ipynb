## Plot the Lines

```python
N = 1000
# a: numpy's ndarray
# get 1,000 points between -100 and 100
a = np.linspace(-100, 100, N)
```

```python
plt.plot(a, a)
```

```python
plt.plot(a, a)
plt.plot(a, np.log10(a))  # overlap onto the previous plotting
```

```python
# zoom in
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.plot(a, a)
plt.plot(a, np.log10(a))
```

```python
x_a = np.linspace(-10, 10, 100)
```

```python
plt.plot(x_a, np.sin(x_a))  # -> range of sin is [-1, 1]
```

```python
two_pi_x_a = np.linspace(0, np.pi*2, 100)
```

```python
plt.ylim((3, -3))
plt.plot(two_pi_x_a, np.sin(two_pi_x_a))  # sin's period is 2π

# frequency: f times in a unit time, 1 Hz = 1 time a second
# period: T unit times for 1 time, f = 1/T
```

```python
plt.ylim((3, -3))
plt.plot(two_pi_x_a, 2*np.sin(two_pi_x_a))  # -> double the range
```

```python
plt.ylim((3, -3))
plt.plot(two_pi_x_a, np.sin(2*two_pi_x_a))  # -> double the frequency
```

## With Noise

```python
N = 1000
signal_a = np.linspace(-100, 100, N)
```

```python
plt.hist(signal_a, bins=15)
```

```python
# loc: = μ = mu = mean
# scale: = σ = sigma = standard deviation
# size: the length of the ndarray
# sample from the normal distribution
noise_a = np.random.normal(loc=0, scale=10, size=N)
```

```python
plt.hist(noise_a, bins=15)
```

```python
sn_a = signal_a + noise_a
```

```python
plt.scatter(x=signal_a, y=sn_a, alpha=0.1)
```

```python
plt.plot(signal_a, signal_a)
plt.scatter(x=signal_a, y=sn_a, alpha=0.1)
```

## Put Texts

```python
N = 1000
a = np.linspace(-100, 100, N)

plt.title('My Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(a, a**2)
plt.text(
    x=90, y=300,
    s=r'$y = x^2$', ha='center', size=15
)
plt.annotate(
    xy=(0, 0),
    arrowprops=dict(facecolor='red', shrink=0.05),
    xytext=(0, 2000),
    s='min', ha='center', size=15,
)
plt.legend(['My Line'])
```
