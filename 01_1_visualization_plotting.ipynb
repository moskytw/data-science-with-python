```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(font='Noto Sans CJK TC')
mpl.style.use('ggplot')  # must put after sns.set
```

## Lines

```python
N = 1000
line_a = np.linspace(-100, 100, N)  # get 1,000 points between -100 and 100
```

```python
plt.plot(line_a, line_a)
```

```python
plt.plot(line_a, np.log(line_a))
```

```python
plt.plot(line_a, line_a)
plt.plot(line_a, np.log(line_a))
```

```python
plt.xlim((0, 10))
plt.ylim((0, 10))
plt.plot(line_a, line_a)
plt.plot(line_a, np.log(line_a))
```

```python
plt.plot(line_a, 10**line_a)
```

```python
plt.yscale('log')
plt.plot(line_a, 10**line_a)
```

```python
plt.plot(line_a, np.sin(line_a))
```

```python
# s: shorter
sline_a = np.linspace(-10, 10, 100)
```

```python
plt.plot(sline_a, np.sin(sline_a))
```

```python
piline_a = np.linspace(0, 2*np.pi, 100)
```

```python
plt.plot(piline_a, np.sin(piline_a))
```

```python
plt.plot(piline_a, 2*np.sin(piline_a))
```

```python
plt.plot(piline_a, np.sin(2*piline_a))
```

## Scatters

```python
noise_a = np.random.normal(loc=0, scale=10, size=N)
```

```python
plt.hist(noise_a, bins=15)
```

```python
sample_with_noise_a = line_a + noise_a
```

```python
plt.scatter(x=line_a, y=sample_with_noise_a, alpha=0.1)
```

```python
plt.plot(line_a, line_a)
plt.scatter(x=line_a, y=sample_with_noise_a, alpha=0.1)
```

## Texts

```python
plt.title('My Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(line_a, line_a**2)
plt.text(
    x=90, y=300,
    s=r'$y = x^2$', ha='center', size=15
)
plt.annotate(
    xy=(0, 0),
    arrowprops=dict(facecolor='red', shrink=0.05),
    xytext=(0, 2000),
    s='min', ha='center', size=15,
)
plt.legend(['My Line'])
```
